```python
import json
import numpy as np

with open('/Users/apple/Desktop/COMP90051/train.json', 'r') as f:
    train_dict = json.load(f)
    
venue_list = []
keywords_list = []
year_list = []
authors_list = []
for i in train_dict.keys():
    if train_dict[i]['venue'] == '':
        venue_list.append('None')
    else:
        venue_list.append(train_dict[i]['venue'])
    if train_dict[i]['keywords'] == '':
        keywords_list.append('None')
    else:
        essay_word_list = []
        for j in train_dict[i]['keywords']:
            essay_word_list.append(str(j))
        keywords_list.append(essay_word_list)
    if train_dict[i]['year'] == '':
        year_list.append('None')
    else:
        year_list.append(train_dict[i]['year'])
    if train_dict[i]['author'] == '':
        authors_list.append('None')
    else:
        essay_author_list = []
        for author in train_dict[i]['author']:
            essay_author_list.append(author)
        authors_list.append(essay_author_list)
```

```python
unique_author = []
for authors in authors_list:
    for author in authors:
        if author not in unique_author:
            unique_author.append(author)

unique_keyword = []
for keywords in keywords_list:
    for keyword in keywords:
        if keyword not in unique_keyword:
            unique_keyword.append(keyword)
```

```python
keywords_array_list = []
for i in train_dict.keys():
    keywords_array = np.zeros(len(unique_keyword))
    for j in train_dict[i]['keywords']:
        keywords_array[j] = 1
    #keywords_array_list.append(keywords_array)
    keywords_array_list.append(keywords_array) 

author_dict = {}
for author in unique_author:
    essay_list = []
    for j in train_dict.keys():
        for co_author in train_dict[j]['author']:
            if author == co_author:
                essay_list.append(j)
    author_dict[author] = essay_list
    
# in the order of author_dict, deal with keywords
keywords_num_array = []
for i in author_dict.keys():
    total_array = np.zeros(len(unique_keyword))
    for j in author_dict[i]:
        number = int(j)
        this_array = keywords_array_list[number]
        total_array += this_array
    keywords_num_array.append(total_array/len(author_dict[i]))
```

```python
small_freq_list = []
for i in range(len(keywords_num_array)):
    small_freq = []
    for j in range(len(keywords_num_array[i])):
        if keywords_num_array[i][j] >= 0.1:
            small_freq.append(j)
    small_freq_list.append(small_freq)
```
